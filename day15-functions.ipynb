```python
def celsius_to_Fahrenheit (temp):
    return(temp*9/5)+32
print(celsius_to_Fahrenheit(0))
print(celsius_to_Fahrenheit (25))
print(celsius_to_Fahrenheit (100))
```

```python
def get_grade(score):
    if score>=90:
        return "A"
    elif score>=75:
        return "B"
    elif score>=50:
        return "C"
    else:
        return "F"

print(get_grade(95))
print(get_grade(82))
print(get_grade(67))
print(get_grade(40))
```
